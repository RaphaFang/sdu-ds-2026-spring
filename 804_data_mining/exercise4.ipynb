### 1. L2, 3*3

```python
q = np.array([
    [0, 4, 0],
    [0, 1, 3],
    [0, 0, 4],
    [1, 3, 0]])

a = np.array([
    [1, 2, 0],
    [0, 1, 2],
    [0, 0, 2],
    [0, 1, 0]])

b = np.array([
    [4, 0, 0],
    [1, 0, 3],
    [0, 0, 4],
    [3, 1, 0]])

c = np.array([
    [1, 1, 2],
    [0, 1, 3],
    [0, 0, 4],
    [1, 2, 1]])

d = np.array([
    [0, 2, 7],
    [0, 0, 3],
    [0, 0, 1],
    [1, 0, 2]])

def process(X, Q):
    dif = X - Q
    the_row_L2 = np.linalg.norm(dif, axis=1)
    the_col_L2 = np.linalg.norm(the_row_L2, axis=0)
    print(the_col_L2)

process(a,q)
process(b,q)
process(c,q)
process(d,q)
```

```python
q_alp = np.array([
    ['W','W','W','W','W','W'],
    ['W','O','O','O','O','W'],
    ['W','O','O','O','R','W'],
    ['W','B','O','B','B','W'],
    ['W','B','B','B','B','W'],
    ['W','W','W','W','W','W'],
])

a_alp = np.array([
    ['W','W','W','W','W','W'],
    ['W','W','R','W','W','W'],
    ['W','O','O','O','W','W'],
    ['B','B','O','B','B','W'],
    ['W','W','W','W','W','W'],
    ['W','W','W','W','W','W'],
])

b_alp = np.array([
    ['W','W','W','W','W','W'],
    ['W','R','R','R','R','W'],
    ['W','R','R','R','O','W'],
    ['W','B','R','B','B','W'],
    ['W','B','B','B','B','W'],
    ['W','W','W','W','W','W'],
])

c_alp = np.array([
    ['W','W','W','W','B','W'],
    ['W','W','O','R','O','W'],
    ['B','O','O','B','O','R'],
    ['W','B','B','O','B','B'],
    ['W','W','B','B','B','W'],
    ['W','W','B','B','W','W'],
])

d_alp = np.array([
    ['W','O','B','B','B','B'],
    ['W','B','O','B','R','R'],
    ['W','B','B','B','B','B'],
    ['W','B','B','B','B','B'],
    ['W','W','W','W','W','W'],
    ['W','W','W','W','W','W'],
])
```

```python
def to_36x3_onehot(X):
    flat = X.reshape(-1) # this will reshape the matrix into (36, )
    out = np.zeros((36, 3)) # this will build a (36, 3) zero m

    out[:, 0] = (flat == "R").astype(int) # masking to get teh count of all the "R"
    out[:, 1] = (flat == "O").astype(int)
    out[:, 2] = (flat == "B").astype(int)
    return out

q_36x3 = to_36x3_onehot(q_alp)
a_36x3 = to_36x3_onehot(a_alp)
b_36x3 = to_36x3_onehot(b_alp)
c_36x3 = to_36x3_onehot(c_alp)
d_36x3 = to_36x3_onehot(d_alp)
```

### 3. L2, 1*1, with "white"


```python
def to_36x4_onehot(X):
    flat = X.reshape(-1)
    out = np.zeros((36, 4))

    out[:, 0] = (flat == "R").astype(int)
    out[:, 1] = (flat == "O").astype(int)
    out[:, 2] = (flat == "B").astype(int)
    out[:, 3] = (flat == "W").astype(int)

    return out

q_36x4 = to_36x4_onehot(q_alp)
a_36x4 = to_36x4_onehot(a_alp)
b_36x4 = to_36x4_onehot(b_alp)
c_36x4 = to_36x4_onehot(c_alp)
d_36x4 = to_36x4_onehot(d_alp)


def process(X, Q):
    dif = X - Q
    the_row_L2 = np.linalg.norm(dif, axis=1)
    the_col_L2 = np.linalg.norm(the_row_L2, axis=0)
    print(the_col_L2)

process(a_36x4,q_36x4)
process(b_36x4,q_36x4)
process(c_36x4,q_36x4)
process(d_36x4,q_36x4)
```

### 4. dist quadratic, 6*6

```python
q = np.array([1, 8, 7])
a = np.array([1, 4, 4])
b = np.array([8, 1, 7])
c = np.array([2, 4, 10])
d = np.array([1, 2, 13])

M_c = np.array([
    [1.0, 0.8, 0.1],
    [0.8, 1.0, 0.0],
    [0.1, 0.0, 1.0]
])
def process_quadratic(X, Q, M_c):
    dif = X - Q
    A = dif @ M_c @ dif.T
    # print(A)

    total = np.sqrt(A)
    print(total)

process_quadratic(a,q, M_c)
process_quadratic(b,q, M_c)
process_quadratic(c,q, M_c)
process_quadratic(d,q, M_c)
```

### 5. dist quadratic, 3*3

```python
q = np.array([
    [0, 4, 0],
    [0, 1, 3],
    [0, 0, 4],
    [1, 3, 0]])

a = np.array([
    [1, 2, 0],
    [0, 1, 2],
    [0, 0, 2],
    [0, 1, 0]])

b = np.array([
    [4, 0, 0],
    [1, 0, 3],
    [0, 0, 4],
    [3, 1, 0]])

c = np.array([
    [1, 1, 2],
    [0, 1, 3],
    [0, 0, 4],
    [1, 2, 1]])

d = np.array([
    [0, 2, 7],
    [0, 0, 3],
    [0, 0, 1],
    [1, 0, 2]])

M_c = np.array([
    [1.0, 0.8, 0.1],
    [0.8, 1.0, 0.0],
    [0.1, 0.0, 1.0]
])
def process_quadratic(X, Q, M_c):
    dif = X - Q
    A = dif @ M_c @ dif.T
    # print(A)
    per_row_sq = np.diag(A)
    total = np.sqrt(per_row_sq.sum())
    print(total)

process_quadratic(a,q, M_c)
process_quadratic(b,q, M_c)
process_quadratic(c,q, M_c)
process_quadratic(d,q, M_c)
```
