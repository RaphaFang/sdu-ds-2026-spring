```python
import numpy as np

# 1. 準備模擬數據 (假設 n=5 筆資料，p=2 個特徵)
# Y: 房價, X1: 坪數, X2: 屋齡
Y = np.array([1000, 1200, 1300, 1500, 1600])
X1 = np.array([30, 40, 50, 60, 70])
X2 = np.array([5, 10, 15, 20, 25])

# ==========================================
# 【對應投影片：The Design Matrix X】
# 為了算出截距(B0)，我們必須在矩陣最左邊塞一排 1
# ==========================================
ones = np.ones(len(Y))
# 把 1, X1, X2 拼成一個大矩陣
X_matrix = np.column_stack((ones, X1, X2)) 

# ==========================================
# 【對應投影片：Least Squares Estimates】
# 終極矩陣公式：β = (X^T * X)^(-1) * X^T * Y
# ==========================================
X_T = X_matrix.T                                  # 1. 取轉置矩陣 (X^T)
X_T_X = X_T.dot(X_matrix)                         # 2. X^T 乘上 X
X_T_X_inv = np.linalg.inv(X_T_X)                  # 3. 取反矩陣 (^-1)
X_T_Y = X_T.dot(Y)                                # 4. X^T 乘上 Y

# 算出來了！這就是包含截距和斜率的 β 陣列
betas = X_T_X_inv.dot(X_T_Y)
print(f"手算的截距 B0: {betas[0]:.2f}")
print(f"手算的坪數斜率 B1: {betas[1]:.2f}")
print(f"手算的屋齡斜率 B2: {betas[2]:.2f}")

# ==========================================
# 【對應投影片：Estimating variance】
# 算殘差平方和 (RSS) 與變異數 (Sigma^2)
# ==========================================
# 1. 算出預測的 Y_hat
Y_hat = X_matrix.dot(betas)
# 2. 算殘差 (真實值 - 預測值)
residuals = Y - Y_hat
# 3. 殘差平方和 RSS = 𝝐^T * 𝝐
RSS = residuals.T.dot(residuals)
# 4. 算變異數 (除以自由度 n - p - 1)
n = len(Y)
p = 2
sigma_squared = RSS / (n - p - 1)
print(f"手算的變異數 Sigma^2: {sigma_squared:.2f}")
```

```python
import numpy as np

# 1. 準備模擬數據 (假設 n=5 筆資料，p=2 個特徵)
# Y: 房價, X1: 坪數, X2: 屋齡
Y = np.array([1000, 1200, 1300, 1500, 1600])
X1 = np.array([30, 40, 50, 60, 70])
# 換成這組數字，打破完美的規律
X2 = np.array([5, 12, 14, 19, 28])

# ==========================================
# 【對應投影片：The Design Matrix X】
# 為了算出截距(B0)，我們必須在矩陣最左邊塞一排 1
# ==========================================
ones = np.ones(len(Y))
# 把 1, X1, X2 拼成一個大矩陣
X_matrix = np.column_stack((ones, X1, X2)) 

# ==========================================
# 【對應投影片：Least Squares Estimates】
# 終極矩陣公式：β = (X^T * X)^(-1) * X^T * Y
# ==========================================
X_T = X_matrix.T                                  # 1. 取轉置矩陣 (X^T)
X_T_X = X_T.dot(X_matrix)                         # 2. X^T 乘上 X
X_T_X_inv = np.linalg.inv(X_T_X)                  # 3. 取反矩陣 (^-1)
X_T_Y = X_T.dot(Y)                                # 4. X^T 乘上 Y

# 算出來了！這就是包含截距和斜率的 β 陣列
betas = X_T_X_inv.dot(X_T_Y)
print(f"手算的截距 B0: {betas[0]:.2f}")
print(f"手算的坪數斜率 B1: {betas[1]:.2f}")
print(f"手算的屋齡斜率 B2: {betas[2]:.2f}")

# ==========================================
# 【對應投影片：Estimating variance】
# 算殘差平方和 (RSS) 與變異數 (Sigma^2)
# ==========================================
# 1. 算出預測的 Y_hat
Y_hat = X_matrix.dot(betas)

# 2. 算殘差 (真實值 - 預測值)
residuals = Y - Y_hat

# 3. 殘差平方和 RSS = 𝝐^T * 𝝐
RSS = residuals.T.dot(residuals)

# 4. 算變異數 (除以自由度 n - p - 1)
n = len(Y)
p = 2
sigma_squared = RSS / (n - p - 1)
print(f"手算的變異數 Sigma^2: {sigma_squared:.2f}")
```

```python
import pandas as pd
import statsmodels.formula.api as smf

# 把剛剛的資料放進 DataFrame
df = pd.DataFrame({'Y': Y, 'X1': X1, 'X2': X2})

# ==========================================
# 【對應投影片：Multiple linear regression - assumptions (R語言報表)】
# 一行程式碼，取代上面所有的矩陣運算！
# ==========================================
model = smf.ols('Y ~ X1 + X2', data=df).fit()

# 印出結果
print(model.summary())
```
