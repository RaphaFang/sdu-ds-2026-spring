```python
import sys
import torch

print("Python executable:", sys.executable)
print("PyTorch version:", torch.__version__)

print("CUDA available:", torch.cuda.is_available())
print("MPS built:", torch.backends.mps.is_built() if hasattr(torch.backends, "mps") else "no mps backend")
print("MPS available:", torch.backends.mps.is_available() if hasattr(torch.backends, "mps") else "no mps backend")

if hasattr(torch.backends, "mps") and torch.backends.mps.is_available():
    device = torch.device("mps")
elif torch.cuda.is_available():
    device = torch.device("cuda")
else:
    device = torch.device("cpu")

print("Using device:", device)

x = torch.randn(2000, 2000).to(device)
y = torch.randn(2000, 2000).to(device)
z = x @ y
print("Result device:", z.device)
```
