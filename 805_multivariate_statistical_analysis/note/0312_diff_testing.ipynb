```python
import numpy as np
import scipy.stats as stats
import pingouin as pg
import pandas as pd
import matplotlib.pyplot as plt

# ==========================================
# 1. 準備模擬資料 (各 500 筆)
# ==========================================
np.random.seed(42) # 固定亂數種子，確保結果一致

# 【資料 A：乖寶寶 MVN】
mean_A = [0, 0]
cov_A = [[1, 0.5], [0.5, 1]] # 有線性相關的橢圓
data_A = np.random.multivariate_normal(mean_A, cov_A, 500)
df_A = pd.DataFrame(data_A, columns=['X', 'Y'])

# 【資料 B：十字架怪物】
X_B = np.random.normal(0, 1, 500)
Z = np.random.choice([-1, 1], 500) # 正負號硬幣
Y_B = X_B * Z
df_B = pd.DataFrame({'X': X_B, 'Y': Y_B})


# ==========================================
# 2. 定義一個「全自動抓妖機」函數
# ==========================================
def run_all_tests(df, dataset_name):
    print(f"\n{'='*40}")
    print(f"🕵️ 正在檢驗：{dataset_name}")
    print(f"{'='*40}")
    
    # --- 一維檢定 (只看單一變數 X 的側臉) ---
    print("\n[一維檢定：K-S Test 與 Shapiro-Wilk]")
    
    ks_stat, ks_pval = stats.kstest(df['X'], 'norm')
    print(f"▶ K-S Test        | p-value: {ks_pval:.4f} " + 
          ("✅ 通過 (保留H0)" if ks_pval > 0.05 else "❌ 妖怪 (拒絕H0)"))
    
    sw_stat, sw_pval = stats.shapiro(df['X'])
    print(f"▶ Shapiro-Wilk    | p-value: {sw_pval:.4f} " + 
          ("✅ 通過 (保留H0)" if sw_pval > 0.05 else "❌ 妖怪 (拒絕H0)"))

    # --- 多維檢定 (看 X 和 Y 合體的 3D 形狀) ---
    print("\n[多維檢定：Henze-Zirkel Test (現代最強把關)]")
    
    # 【修正點】：直接把整個 dataframe 丟進去就好，不用加 method 參數！
    hz_result = pg.multivariate_normality(df)
    
    print(f"▶ Henze-Zirkel    | p-value: {hz_result.pval:.4f} " + 
          ("✅ 完美 MVN" if hz_result.normal else "❌ 聯合妖怪!"))

# ==========================================
# 3. 讓兩組資料上處刑台
# ==========================================
run_all_tests(df_A, "資料 A (乖寶寶 MVN 橢圓)")
run_all_tests(df_B, "資料 B (十字架怪物)")
```
